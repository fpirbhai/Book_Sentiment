# Load the book

```python
with open('miracle_in_the_andes.txt', 'r', encoding='utf-8') as file:
    book = file.read()
```

```python
type(book)
```

# How many chapters?

Using String Method

```python
book.count('Chapter')
```

With Regex    

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]+")
pattern
```

```python
findings = re.findall(pattern=pattern, string=book)
findings
```

```python
len(findings)
```

# Which are the sentences where the word 'love' was used?

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z][lL]ove[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern = pattern, string = book)
```

```python
findings[0]
```

```python
len(findings)
```

# What are the most used words

```python
book.split(" ")
```

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book)
findings
```

```python
dict_word = {}

for word in findings:
    if word in dict_word.keys():
        dict_word[word] = dict_word[word] + 1
    else:
        dict_word[word] = 1
```
